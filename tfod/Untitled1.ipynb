```python
import albumentations as A
import os
import imageio
import numpy as np
output_path="C:/Users/Harsh/Desktop/project industry/Original images/augmented images/"
input_path="C:/Users/Harsh/Desktop/project industry/Original images/rescaled images/"
# import pathlib
# output_path=pathlib.Path(output_path)
# input_path=pathlib.Path(input_path)
```

```python
input_images=os.listdir(input_path)
input_images.sort()
```

```python
def read_image(url):
    image=imageio.imread(url)
    return image

def augemented_and_return(aug,image):
    aimg=aug(image=image)['image']
    return aimg

def save_augmented_image(aimg,image,directory):
    uri=directory+"/"+image
    imageio.imwrite(uri,aimg)
```

```python
listagF={
"VerticalFlip":"A.VerticalFlip(p=1)",
"HorizontalFlip":"A.HorizontalFlip(p=1)",
"Rotate1":"A.Rotate(limit=90)",
"Rotate2":"A.Rotate(limit=-90)",
"Rotate3":"A.Rotate(limit=180)",
"ImageCompression":"A.ImageCompression(quality_lower=0,quality_upper=1)",
"InvertImg":"A.InvertImg(p=1)",
"MultiplicativeNoise1":"A.MultiplicativeNoise(p=0.8)",
"MultiplicativeNoise2":"A.MultiplicativeNoise(p=1)",
"GaussNoise":"A.GaussNoise(p=1)",
"GaussianBlur":"A.GaussianBlur(p=1)",
"RandomCrop":"A.RandomCrop(height=400,width=400,p=1)"}
```

```python
keys=list(listagF.keys())
values=list(listagF.values())
```

```python
for k in range(len(keys)):
    print(keys[k])
    
    aug=eval(values[k])
    
    folder=keys[k]
    
    for i in input_images:
        print(i)
        path2img=input_path+i
        image=read_image(path2img)
        
        aimg=augemented_and_return(aug,image)
        
        dest=output_path+folder+"/"
        
        if not os.path.exists(dest):
            os.mkdir(dest)
        
        save_augmented_image(aimg,i,dest)
```
